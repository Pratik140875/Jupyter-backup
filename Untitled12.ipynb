```python
import pandas as pd
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\prati\Downloads\customer_churn.csv')
df.sample(5)
```

```python
df = df.drop(['customerID'], axis = 'columns')
df.head()
```

```python
df.dtypes
```

```python
df1 = df[df['TotalCharges']!=' ']
df1.head()
```

```python
df1.TotalCharges = pd.to_numeric(df.TotalCharges, errors = 'coerce')
df1.head()
```

```python
df1.dtypes
```

```python
for col in df1:
    print(f'{col} : {df1[col].unique()}')
```

```python
df1.replace({"No internet service": "No"}, inplace = True)
df1.replace({"No phone service": "No"}, inplace = True)
```

```python
for col in df1:
    print(f'{col} : {df1[col].unique()}')
```

```python
colum = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']
for i in colum:
    df1[i].replace({'Yes':1, 'No':0}, inplace = True)
```

```python
df1.head()
```

```python
for col in df1:
    print(f'{col}:{df1[col].unique()}')
```

```python
df1['OnlineBackup'].replace({'Yes':1,'No':0}, inplace = True)
```

```python
for col in df1:
    print(f'{col}:{df1[col].unique()}')
```

```python
df1['gender'].replace({'Female':1, 'Male':0}, inplace = True)
```

```python
for col in df1:
    print(f'{col}:{df1[col].unique()}')
```

```python
df2 = pd.get_dummies(data = df1, columns = ['InternetService', 'Contract', 'PaymentMethod'])
df2.columns
```

```python
df2.sample(5)
```

```python
col_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[col_to_scale] = scaler.fit_transform(df2[col_to_scale])
```
